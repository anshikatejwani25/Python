```python
import datetime

# Dictionaries of available product and their prices
products={
    "apple"       : 30,
    "banana"      : 10,
    "orange"      : 25,
    "mango"       : 60,
     "potato"     : 20,
    "tomato"      : 25,
    "onion"       : 30,
    "carrot"      : 35,
    "milk"        : 50,
    "bread"       : 40,
    "butter"      : 55,
    "cheese"      : 90,
    "rice"        : 70,
    "wheat flour" : 45,
    "sugar"       : 40,
    "salt"        : 20,
    "oil"         : 120,
    "tea"         : 150,
    "coffee"      : 200,
    "chocolate"   : 45,
    "chips"       : 30,
    "biscuits"    : 25,
    "soap"        : 25,
    "shampoo"     : 90,
    "toothpaste"  : 45,
    "toothbrush"  : 20,
    "handwash"    : 35
}

bill_number = 1
#Executes for every new Customer
print("-"*50)
print("Welcome to the Supermarket".center(50))
print("-"*50)

#Main loop 
while True:
    amount=0
    items=[] #To store purchased items
    name=input("Enter name: ")
    current_datetime = datetime.datetime.now().strftime("%d-%m-%Y %H:%M:%S")
    phone=input("Enter phone_no: ")

#Ask for membership status
    membership = input("Are you a member? (yes/no): ").lower()

    print("Enter product name and quantity")

# Inner loop for processing items
    while True:
        p_name=input("Enter Product_name: ").lower()
 
        if p_name not in products:
            print("Item not found.Try again.")
            continue

        quantity=float(input("Enter Quantity: "))
        price=products[p_name] #price from dictionary
        total_price= price*quantity
        amount+= total_price
        
#Store items in a list
        items.append([p_name,quantity,price,total_price])
        
#Ask to add more items
        next_item= input("Do you want to add more items (yes/no): ")
        if next_item=="no":
            break

#Apply discount based on total 
    if amount <= 500:
        discount = 0.05
    elif amount <= 1000:
        discount = 0.07
    elif amount <= 5000:
        discount = 0.15
    else:
        discount=0.2

    if membership == "yes":
        discount += 0.05  # extra 5% for members
            
    discount_amount = discount*amount
    Final_amount = amount - discount_amount

# GST calculation (18%)
    gst = Final_amount * 0.18
    Final_amount += gst

#Print Bill
    print("\n"+"-"*50)
    print("Supermarket Bill".center(50))
    print("-"*50)
    print(f"Bill No: {bill_number}")
    print(f"Date & Time: {current_datetime}")
    print(f"Customer_name: {name}")
    print(f"Phone_number: {phone}")
    print(f"Membership: {'Yes' if membership == 'yes' else 'No'}")
    print("-"*50)
    
#Table Header
    print(f"{'Items':15}{'Qty':>8}{'Price':>10}{'Total':>12}")
    print("-"*50)
    
#Purchased Item
    for item in items:
        print(f"{item[0]:15}{item[1]:>8}{item[2]:>10}{item[3]:>12}")

    print("-"*50)
    print(f"{'Subtotal':>35} : {amount:.2f}")
    print(f"{'Discount':>35} : {discount_amount:.2f}")
    print(f"{'gst (18%)':>35} : {gst:.2f}")
    print(f"{'Final_Amount':>35} : {Final_amount:.2f}")
    print("-"*50)
    print("Thank You For Shopping With Us!\n".center(50))

    bill_number += 1
    
#Ask to add another Customer
    Next_Cust=input("Do you want to serve next customer? (yes/no): ").lower()
    if  Next_Cust != "yes":
        break
```
